```python

AI_FOUNDRY_PROJECT_ENDPOINT= "https://idkopenai.services.ai.azure.com"
AI_FOUNDRY_DEPLOYMENT_NAME=  "gpt-4o"
embedding_model = "text-embedding-ada-002"

AI_FOUNDRY_API_VERSION= "2024-12-01-preview"
AI_FOUNDRY_API_KEY= "" 
```

```python
import openai

# Create a client using your Azure OpenAI endpoint and API key
client = openai.AzureOpenAI(
    api_key= AI_FOUNDRY_API_KEY,
    azure_endpoint= AI_FOUNDRY_PROJECT_ENDPOINT,
    api_version= AI_FOUNDRY_API_VERSION
)

deployment_name = AI_FOUNDRY_DEPLOYMENT_NAME  # e.g., "gpt-4o" or your custom deployment

# Send a chat completion request
response = client.chat.completions.create(
    model=deployment_name,  # For Azure, this is your deployment name
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Hello, who won the world cup in 2018?"}
    ],
    max_tokens=100,
    temperature=0.7
)

# Print the AI's reply
print(response.choices[0].message.content)
```
